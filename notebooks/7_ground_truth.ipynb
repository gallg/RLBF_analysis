```python
import numpy as np
```

### Define ground truth for frequency (Simulation)

```python
# Calculate simulation reward assuming that 
# contrast (c) is linear;
c = 1 
f = 7
reward = []
for f in np.arange(0,10,0.1):
    reward.append(c * np.exp(-((f-7)/3) ** 2))

import matplotlib.pyplot as plt
plt.plot(np.arange(0,10,0.1), reward)
```
